```python
# Using the quotes API, build a GUI that provides a drop-down menu of people randomly chosen from the API
```

```python
import requests
from tkinter import *
import random

counter = 0
list_authors = []

authors = requests.get("https://api.quotable.io/authors?sortBy=name")
authors = authors.json()

for x in range(20):
    x = authors['results'][counter]['name']
    list_authors.append(x)
    counter += 1
    
print(list_authors)


root = Tk()
root.title("Random Quote Generator")
root.geometry("600x600")

def choose_author():
    author = value.get()
    result_label = Label(root, text=requests.get(f"https://api.quotable.io/random?author={author}").text)
    result_label.pack()
    
value = StringVar()
value.set("Select an author: ")


three_random = []
for x in range(3):
    rand_auth = random.choice(list_authors)
    three_random.append(rand_auth)

dropdown = OptionMenu(root, value, *three_random)
choose_button = Button(root, text="Click To Generate A Random Quote", command=choose_author)

dropdown.pack()
choose_button.pack()

root.mainloop()
```
